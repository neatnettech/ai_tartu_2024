```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from math import pi

data = {
    'composite_key': ['49.0 MALE', '26.0 FEMALE', '29.0 FEMALE'],
    'bars_audible_adu_q_1_1c_q_1': [2.0, 2.0, 2.0],
    'celltowers_health_adu_q_1_1c_q_1': [4.0, 3.0, 4.0],
    'neighbours_audible_adu_q_1_1c_q_1': [2.0, 1.0, 1.0],
    'planes_hindrance_adu_q_1_1c_q_1': [0.0, 1.0, 8.0],
    'powerlines_health_adu_q_1_1c_q_1': [5.0, 3.0, 4.0],
    'sleep_morning_adu_q_1_1c_q_1': [3.0, 4.0, 3.0],
    'trains_audible_adu_q_1_1c_q_1': [0.0, 1.0, 0.0],
}

hospitalization_data = {
    'composite_key': ['49.0 MALE', '26.0 FEMALE', '29.0 FEMALE'],
    'hospitalization_pastyear_adu_q_1_a_1c_q_1': [60.0, 120.0, 100.0],
}

df_top3 = pd.DataFrame(data)
df_hospitalization = pd.DataFrame(hospitalization_data)

categories = list(df_top3.columns[1:])
N = len(categories)

angles = [n / float(N) * 2 * pi for n in range(N)]
angles += angles[:1]  # Complete the loop

plt.figure(figsize=(8, 8))
ax = plt.subplot(111, polar=True)

for i in range(len(df_top3)):
    values = df_top3.loc[i].drop('composite_key').values.flatten().tolist()
    values += values[:1]  # Complete the loop
    ax.plot(angles, values, label=df_top3['composite_key'][i], linewidth=2)
    ax.fill(angles, values, alpha=0.25)

plt.xticks(angles[:-1], categories, color='grey', size=10)
ax.set_rlabel_position(0)
plt.yticks([0, 1, 2, 3, 4, 5, 6, 7, 8], ["0", "1", "2", "3", "4", "5", "6", "7", "8"], color="grey", size=7)
plt.ylim(0, 8)

plt.title('Radar Chart of Sleep-Related Problems (Excluding Hospitalization)', size=15, color='black', y=1.1)
plt.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))

plt.tight_layout()
plt.show()

plt.figure(figsize=(8, 6))
plt.bar(df_hospitalization['composite_key'], df_hospitalization['hospitalization_pastyear_adu_q_1_a_1c_q_1'], color=['red', 'blue', 'green'])
plt.title('Hospitalization Data for Top 3 Individuals', size=15)
plt.xlabel('Individuals')
plt.ylabel('Hospitalization (Past Year)')
plt.show()
```
